# generating demo data

```python
from pathlib import Path
import pandas as pd
```

```python
demo_data = []
n_records = 500 #getting 20 records for each label in each plate
```

```python
# coloc model

data_path = Path("/project/restored-files/Gurvir/Saved_ConcatData/")
data_path.is_dir()

out_path = Path("/project/restored-files/kdsa/paper/disease_classifier_nat_res/demo_data/")
```

```python
df = pd.read_csv(data_path.joinpath(f"CombinedDataCleaned.csv"), index_col=0)
print(df.shape)
display(df.head(3))
```

```python
plates = df.PlateID.unique() 
print(plates)
```

```python
df.groupby(['PlateID'])['label_name'].unique()
```

```python


for plate in plates:
    #print("Plate", plate)
    l_df_plate = df.loc[df['PlateID'] == plate]
    l_df = l_df_plate.groupby('label_name').head(n_records)
    demo_data.append(l_df)
    del l_df_plate, l_df
    
```

```python
demo_data = pd.concat(demo_data, axis=0)
```

```python
print(demo_data.shape)
demo_data.groupby(['PlateID'])['label_name'].unique()
```

```python
#output
print(out_path.joinpath('coloc_tab_demo_data.csv'))
demo_data.to_csv(out_path.joinpath('coloc_tab_demo_data.csv'))
```
